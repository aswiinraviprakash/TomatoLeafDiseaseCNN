```python
import numpy as np
import matplotlib.pyplot as plt

dataset3_metrics = np.load('mobilenet_dataset_3_metrics.npz')
tl_dataset3_metrics = np.load('mobilenet_transferLearning_dataset_3_metrics.npz')

dataset3_train_loses = dataset3_metrics['train_losses']
dataset3_validate_loses = dataset3_metrics['validate_losses']

tl_dataset3_train_loses = tl_dataset3_metrics['train_losses']
tl_dataset3_validate_loses = tl_dataset3_metrics['validate_losses']

plt.plot(dataset3_train_loses,label='MobileNetV2 Train Loss', linestyle='dashed')
plt.plot(dataset3_validate_loses,label='MobileNetV2 Validate Loss')
plt.plot(tl_dataset3_train_loses,label='MobileNetV2 Transfer Learning Train Loss', linestyle='dashed')
plt.plot(tl_dataset3_validate_loses,label='MobileNetV2 Transfer Learning Validate Loss')
plt.title('Loss Comparison Without and With Transfer Learning')
plt.legend()
plt.show()
```

```python
dataset3_train_accuracies = dataset3_metrics['train_accuracies']
dataset3_validate_accuracies = dataset3_metrics['validate_accuracies']

tl_dataset3_train_accuracies = tl_dataset3_metrics['train_accuracies']
tl_dataset3_validate_accuracies = tl_dataset3_metrics['validate_accuracies']

plt.plot(dataset3_train_accuracies,label='MobileNetV2 Train Accuracy', linestyle='dashed')
plt.plot(dataset3_validate_accuracies,label='MobileNetV2 Validate Accuracy')
plt.plot(tl_dataset3_train_accuracies,label='MobileNetV2 Transfer Learning Train Accuracy', linestyle='dashed')
plt.plot(tl_dataset3_validate_accuracies,label='MobileNetV2 Transfer Learning Validate Accuracy')
plt.title('Accuracy Comparison Without and With Transfer Learning')
plt.legend()
plt.show()
```

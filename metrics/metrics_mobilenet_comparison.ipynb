```python
import numpy as np
import matplotlib.pyplot as plt

dataset1_metrics = np.load('mobilenet_dataset_1_metrics.npz')
dataset2_metrics = np.load('mobilenet_dataset_2_metrics.npz')
dataset3_metrics = np.load('mobilenet_dataset_3_metrics.npz')

dataset1_train_loses = dataset1_metrics['train_losses']
dataset1_validate_loses = dataset1_metrics['validate_losses']

dataset2_train_loses = dataset2_metrics['train_losses']
dataset2_validate_loses = dataset2_metrics['validate_losses']

dataset3_train_loses = dataset3_metrics['train_losses']
dataset3_validate_loses = dataset3_metrics['validate_losses']

plt.plot(dataset1_train_loses,label='Dataset 1 Train Loss', linestyle='dashed')
plt.plot(dataset1_validate_loses, label='Dataset 1 Validate Loss')
plt.plot(dataset2_train_loses,label='Dataset 2 Train Loss', linestyle='dashed')
plt.plot(dataset2_validate_loses,label='Dataset 2 Validate Loss')
plt.plot(dataset3_train_loses,label='Dataset 3 Train Loss', linestyle='dashed')
plt.plot(dataset3_validate_loses,label='Dataset 3 Validate Loss')
plt.title('MobileNetV2 Train and Validate Loss Comparison All Models')
plt.legend()
plt.show()
```

```python
dataset1_train_accuracies = dataset1_metrics['train_accuracies']
dataset1_validate_accuracies = dataset1_metrics['validate_accuracies']

dataset2_train_accuracies = dataset2_metrics['train_accuracies']
dataset2_validate_accuracies = dataset2_metrics['validate_accuracies']

dataset3_train_accuracies = dataset3_metrics['train_accuracies']
dataset3_validate_accuracies = dataset3_metrics['validate_accuracies']

plt.plot(dataset1_train_accuracies,label='Dataset 1 Train Accuracy', linestyle='dashed')
plt.plot(dataset1_validate_accuracies, label='Dataset 1 Validate Accuracy')
plt.plot(dataset2_train_accuracies,label='Dataset 2 Train Accuracy', linestyle='dashed')
plt.plot(dataset2_validate_accuracies,label='Dataset 2 Validate Accuracy')
plt.plot(dataset3_train_accuracies,label='Dataset 3 Train Accuracy', linestyle='dashed')
plt.plot(dataset3_validate_accuracies,label='Dataset 3 Validate Accuracy')
plt.title('MobileNetV2 Train and Validate Accuracy Comparison All Models')
plt.legend()
plt.show()
```

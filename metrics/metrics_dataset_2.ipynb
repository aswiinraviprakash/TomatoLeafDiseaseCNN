```python
import numpy as np
import matplotlib.pyplot as plt

resnet_metrics = np.load('resnet_dataset_2_metrics.npz')
mobilenet_metrics = np.load('mobilenet_dataset_2_metrics.npz')
vgg_metrics = np.load('vgg_dataset_2_metrics.npz')

resnet_train_loses = resnet_metrics['train_losses']
resnet_validate_loses = resnet_metrics['validate_losses']

vgg_train_loses = vgg_metrics['train_losses']
vgg_validate_loses = vgg_metrics['validate_losses']

mobilenet_train_loses = mobilenet_metrics['train_losses']
mobilenet_validate_loses = mobilenet_metrics['validate_losses']


plt.plot(resnet_train_loses,label='ResNet18 Train Loss', linestyle='dashed')
plt.plot(resnet_validate_loses, label='ResNet18 Validate Loss')
plt.plot(vgg_train_loses,label='VGG16 Train Loss', linestyle='dashed')
plt.plot(vgg_validate_loses,label='VGG16 Validate Loss')
plt.plot(mobilenet_train_loses,label='MobileNetV2 Train Loss', linestyle='dashed')
plt.plot(mobilenet_validate_loses,label='MobileNetV2 Validate Loss')
plt.title('Train and Validate Loss Comparison')
plt.legend()
plt.show()
```

```python
resnet_train_accuracies = resnet_metrics['train_accuracies']
resnet_validate_accuracies = resnet_metrics['validate_accuracies']

vgg_train_accuracies = vgg_metrics['train_accuracies']
vgg_validate_accuracies = vgg_metrics['validate_accuracies']

mobilenet_train_accuracies = mobilenet_metrics['train_accuracies']
mobilenet_validate_accuracies = mobilenet_metrics['validate_accuracies']


plt.plot(resnet_train_accuracies,label='ResNet18 Train Accuracy', linestyle='dashed')
plt.plot(resnet_validate_accuracies, label='ResNet18 Validate Accuracy')
plt.plot(vgg_train_accuracies,label='VGG16 Train Accuracy', linestyle='dashed')
plt.plot(vgg_validate_accuracies,label='VGG16 Validate Accuracy')
plt.plot(mobilenet_train_accuracies,label='MobileNetV2 Train Accuracy', linestyle='dashed')
plt.plot(mobilenet_validate_accuracies,label='MobileNetV2 Validate Accuracy')
plt.title('Train and Validate Accuracy Comparison')
plt.legend()
plt.show()
```
